# Creating Movie Recommendation

```python
#Create Matrix
moviemat = df.pivot_table(index="User_id",columns="Title",values="Rating")
moviemat.head()
```

```python
# sort the Movies on the basis of highest Value of Ratings
Ratings.sort_values('num of Ratings',ascending=False)
```

```python
#Now lets pick any movie out from the list and check it's user wise Ratings
# suppose we take starwars
starwars_user_ratings=moviemat['Star Wars (1977)']
starwars_user_ratings.head()
```

```python
#Now corelate the series with the main dataframe of moviematrix
similar_to_starwars=moviemat.corrwith(starwars_user_ratings)
similar_to_starwars
```

```python
# convert this series into DataFrames
corr_starwars=pd.DataFrame(similar_to_starwars,columns=['correlation'])
corr_starwars
```

```python
#remove NaN value
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars
```

```python
# join corr_starwars and Ratings
corr_starwars=corr_starwars.join(Ratings['num of Ratings'])
corr_starwars
```

```python
# we have to Take only those movies whose Num of Ratings is more than 100
corr_starwars[corr_starwars['num of Ratings']>100].sort_values('correlation',ascending=False)
```

# PREDICT FUNCTION

```python
def predict_movie(movie_name):
    movie_user_ratings=moviemat[movie_name]
    similar_to_movie=moviemat.corrwith(movie_user_ratings)
    corr_movie=pd.DataFrame(similar_to_movie,columns=['correlations'])
    corr_movie.dropna(inplace=True)
    #join
    corr_movie=corr_movie.join(Ratings['num of Ratings'])
    # sorting
    predictions=corr_movie[corr_movie['num of Ratings']>100].sort_values('correlations',ascending=False)
    return predictions
    
```

```python
pred=predict_movie("Titanic (1997)")
pred.head()
```
