# HISTOGRAM

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# TO GENERATE SOME DATA USING NORMAL DISTRIBUTION
xsn=np.random.randn(100)
sigma=5
U=60
x=np.round(xsn*sigma+U)
x2=np.round(xsn*5+40)
print(x)
print(x2)
```

```python
#PLOT A HISTOGRAM
plt.style.use("seaborn")
plt.xlabel("prob/frecount of student")
plt.ylabel("marks Range")
plt.hist(x,alpha=0.8)
plt.hist(x2,alpha=0.8)
plt.show()
```
