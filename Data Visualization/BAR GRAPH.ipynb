# BAR GRAPH

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_coordinates=np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15],width=0.5,label="current year",tick_label=["gold","platinum","silver"])
plt.bar(x_coordinates+0.25,[20,10,12],width=0.5,label="next year")
plt.title("metal price comparison")
plt.xlabel("metal")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
plt.style.use("dark_background")
x_coordinates=np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15],width=0.5,label="current year",tick_label=["gold","platinum","silver"])
plt.bar(x_coordinates+0.25,[20,10,12],width=0.5,label="next year")
#Limit
plt.ylim(0,40)
plt.xlim(-2,5)
plt.title("metal price comparison")
plt.xlabel("metal")
plt.ylabel("price")
plt.legend()
plt.show()
```
