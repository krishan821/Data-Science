# LINE PLOT


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(10)
y1=x**x
y2=2*x+3
print(x)
print(y1)
print(y2)
```

```python
themes=plt.style.available
print(themes)
```

```python
plt.style.use('seaborn-notebook')
```

```python
plt.plot(x,y1,color='red',label='apple',linestyle='dotted',marker='o')
plt.plot(x,y2,color='green',label='kiwi',linestyle='dotted',marker='*')
plt.xlabel('time')
plt.ylabel('price')
plt.title('price of fruits over time')
plt.legend()
plt.show()
```
