```python
import seaborn as sns
import numpy as np
```

```python
tips_data=sns.load_dataset('tips')
```

```python
tips_data
```

# BAR PLOT


```python
barplot=sns.barplot(x='sex',y='tip',data=tips_data)
print(barplot)
```

```python
# Select the row which has sex = female
tips_data[tips_data["sex"]=="Female"]["tip"].mean()
```

```python
#plot barplot on basis of other statistics property
sns.barplot(x="sex",y="tip",data=tips_data,estimator=np.std)
```

# Count Plot

```python
sns.countplot(x="sex",data=tips_data)
```

```python
sns.countplot(x="day",data=tips_data)
```

```python
# using Hue it help us to segregate the feature into two component/ it work with multiple component
sns.countplot(x="day",hue="sex",data=tips_data)
```

# Box Plot


```python
sns.boxplot(x='sex',y='tip',data=tips_data)
```

```python
sns.boxplot(x='day',y='tip',hue='sex',data=tips_data)
```

# Violin Plot

```python
sns.violinplot(x='day',y='tip',hue='sex',data=tips_data)
```

# Distribution Plot

```python
 sns.distplot(tips_data['tip'])
```

```python
# Remove estimation
sns.distplot(tips_data['tip'],kde=False)
```

# KDE Plot

```python
sns.kdeplot(tips_data['tip'])
```

# Joint Plot

```python
sns.jointplot(x="total_bill",y="tip",data=tips_data,kind="scatter")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips_data,kind="hex")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips_data,kind="kde")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips_data,kind="reg")
```

# Pair Plot

```python
sns.pairplot(tips_data)
```

```python
sns.pairplot(tips_data,hue="sex")
```

# Heatmap

```python
flights=sns.load_dataset("Flights")
flights
```

```python
flights_pivot=flights.pivot_table(index="month",columns="year",values="passengers")
flights_pivot
```

```python
sns.heatmap(flights_pivot)
```

```python
sns.heatmap(tips_data.corr())
```

```python
sns.heatmap(tips_data.corr(),cmap="coolwarm")
```
